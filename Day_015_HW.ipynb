```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = np.random.uniform(low=-1, high=1, size=(10,10))
plt.figure(figsize=(10,10))
"""
Your Code Here
"""
sns.heatmap(matrix, cmap= plt.cm.RdYlBu_r, vmin = -0.25, annot = True)
plt.show(sns.heatmap)

# heatmap參數
# vmax,vmin: 分別是熱力圖的顏色取值最大最小範圍, 預設為資料表的取值
# camp: 表示數字到色彩空間的對映.可由matplotlib包中的colormap名稱或是顏色物件取值. 若想更改預設值可以根據center引數設定.
# annot(annotate的縮寫): 預設值為False. 如果為True, 在熱力圖中方格內寫入資料.
# fmt: 矩陣上標示數字的資料格式, Ex.保留小數點後幾位數字
# annot_kws: 預設為False. 如果為True, 可以設定矩陣上數字的大小顏色字型, 須根據matplotlib中的text參數設定.
# linewidths: 熱力圖中"兩兩特徵關係的矩陣小圖"之間的間隔大小
# linecolor: 切分熱力圖上每個小矩陣的線顏色, 預設為白色
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.uniform(low=-1, high=1, size=(nrow, ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()

# matrix = np.random.randn(nrow * ncol).reshape((nrow, ncol)) 
# reshape: 把向量轉變成矩陣
```
